## Hielke Experimentation

```python

def label_precision(row):
    if row['interviewee (migrant/source person)'] == 'migrant':
        return row['precision migrant (migrants, asylum seeker, refugee)']
    if row['interviewee (migrant/source person)'] == 'source person':
        return row['precision source person']
    return pd.NA

new_df['precision_interviewee'] = new_df.apply(lambda row: label_precision(row), axis=1)
new_df.dropna(axis=0, how='all', inplace=True)
```

```python
temp_df = []
for row in new_df.itertuples(index=False):
    if row[2] > 1:
        temp_df.extend([list(row)]*int(row[2]))
    else:
        temp_df.append(list(row))
altered_df = pd.DataFrame(temp_df, columns=new_df.columns)
```

```python
def label_ngo_IOM(row):
    if row['Name of NGO or other who helped'] == 'IOM' or row['Name of NGO who helped 2'] == 'IOM' or row['Name of NGO who helped 3'] == 'IOM':
        return 1
    return 0

def label_ngo_citizens(row):
    if str(row['Name of NGO or other who helped']).startswith('citizen') or str(row['Name of NGO who helped 2']).startswith('citizen') or str(row['Name of NGO who helped 3']).startswith('citizen'):
        return 1
    return 0

def label_ngo_UNHCR(row):
    if row['Name of NGO or other who helped'] == 'UNHCR' or row['Name of NGO who helped 2'] == 'UNHCR' or row['Name of NGO who helped 3'] == 'UNHCR':
        return 1
    return 0

def label_ngo_conseil(row):
    if str(row['Name of NGO or other who helped']).startswith('Conseil Tunisien pour') or str(row['Name of NGO who helped 2']).startswith('Conseil Tunisien pour') or str(row['Name of NGO who helped 3']).startswith('Conseil Tunisien pour'):
        return 1
    return 0

def label_ngo_cellule(row):
    if str(row['Name of NGO or other who helped']).startswith('Cellule') or str(row['Name of NGO who helped 2']).startswith('Cellule') or str(row['Name of NGO who helped 3']).startswith('Cellule'):
        return 1
    return 0

def label_ngo_other(row):
    if not pd.isnull(row['Name of NGO or other who helped']):
        if row.helped_by_IOM + row.helped_by_citizens + row.helped_by_UNCHR + row.helped_by_conseil + row.helped_by_cellule < 1:
            return 1
    return 0

def label_ngo_none(row):
    if pd.isnull(row['Name of NGO or other who helped']):
        return 1
    return 0

altered_df['helped_by_IOM'] = altered_df.apply(lambda row: label_ngo_IOM(row), axis=1)
altered_df['helped_by_citizens'] = altered_df.apply(lambda row: label_ngo_citizens(row), axis=1)
altered_df['helped_by_UNCHR'] = altered_df.apply(lambda row: label_ngo_UNHCR(row), axis=1)
altered_df['helped_by_conseil'] = altered_df.apply(lambda row: label_ngo_conseil(row), axis=1)
altered_df['helped_by_cellule'] = altered_df.apply(lambda row: label_ngo_cellule(row), axis=1)
altered_df['helped_by_other_organization'] = altered_df.apply(lambda row: label_ngo_other(row), axis=1)
altered_df['not_helped_by_organization'] = altered_df.apply(lambda row: label_ngo_none(row), axis=1)
   
```

```python
def label_help_economical(row):
    if row['received help  (yes economical, yes sanitary, yes but not enough, no)'] == 'yes economical' or row['received help (yes economical, yes sanitary, yes but not enough, no)'] == 'yes economical' or row['received help  (yes economical, yes sanitary, yes but not enough, no).1'] == 'yes economical':
        return 1
    return 0

def label_help_sanitary(row):
    if row['received help  (yes economical, yes sanitary, yes but not enough, no)'] == 'yes sanitary' or row['received help (yes economical, yes sanitary, yes but not enough, no)'] == 'yes sanitary' or row['received help  (yes economical, yes sanitary, yes but not enough, no).1'] == 'yes sanitary':
        return 1
    return 0

def label_help_not_enough(row):
    if row['received help  (yes economical, yes sanitary, yes but not enough, no)'] == 'yes but not enough' or row['received help (yes economical, yes sanitary, yes but not enough, no)'] == 'yes but not enough' or row['received help  (yes economical, yes sanitary, yes but not enough, no).1'] == 'yes but not enough':
        return 1
    return 0

def label_help_no(row):
    if row['received help  (yes economical, yes sanitary, yes but not enough, no)'] == 'no' or row['received help (yes economical, yes sanitary, yes but not enough, no)'] == 'no' or row['received help  (yes economical, yes sanitary, yes but not enough, no).1'] == 'no':
        return 1
    return 0

altered_df['received_economical_help'] = altered_df.apply(lambda row: label_help_economical(row), axis=1)
altered_df['received_sanitary_help'] = altered_df.apply(lambda row: label_help_sanitary(row), axis=1)
altered_df['received_not_enough_help'] = altered_df.apply(lambda row: label_help_not_enough(row), axis=1)
altered_df['received_no_help'] = altered_df.apply(lambda row: label_help_no(row), axis=1)
```

```python
altered_df['Ill (yes/no)'] = altered_df['Ill (yes/no)'].map({'Yes': 1, 'No': 0})
altered_df['Ill (yes/no)'].fillna(0, inplace=True)
```

```python
altered_df['Ill (yes/no)'] = altered_df['Ill (yes/no)'].astype('int')
```
