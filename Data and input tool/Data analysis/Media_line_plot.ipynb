# All countries top 5 and top 10 of Country that was reported on

```python
top_10_most_frequent = df['Event Type'].value_counts()[:10].index.tolist()
top_5_most_frequent = df['Event Type'].value_counts()[:5].index.tolist()
```

```python
top_10_filtered_df = df[df['Event Type'].isin(top_10_most_frequent)]
top_5_filtered_df = df[df['Event Type'].isin(top_5_most_frequent)]
```

```python
top_10_filtered_df['Event Date'] = pd.to_datetime(top_10_filtered_df['Event Date'], errors='coerce')
top_10_filtered_df = top_10_filtered_df[top_10_filtered_df['Latitude-city1_degrees'].notna()]
top_10_filtered_df = top_10_filtered_df[top_10_filtered_df['Event Date'] > '2020-01-01']

top_5_filtered_df['Event Date'] = pd.to_datetime(top_5_filtered_df['Event Date'], errors='coerce')
top_5_filtered_df = top_5_filtered_df[top_5_filtered_df['Latitude-city1_degrees'].notna()]
top_5_filtered_df = top_5_filtered_df[top_5_filtered_df['Event Date'] > '2020-01-01']
```

```python
top_10_df_gb = top_10_filtered_df.groupby([top_10_filtered_df['Event Date'].dt.year.rename('year'), top_10_filtered_df['Event Date'].dt.month.rename('month'), top_10_filtered_df['Event Type']]).size().unstack(level=2)
top_10_df_gb.fillna(0, inplace=True)

top_5_df_gb = top_5_filtered_df.groupby([top_5_filtered_df['Event Date'].dt.year.rename('year'), top_5_filtered_df['Event Date'].dt.month.rename('month'), top_5_filtered_df['Event Type']]).size().unstack(level=2)
top_5_df_gb.fillna(0, inplace=True)
```

```python
top_5_df_gb_res = top_5_df_gb.div(top_5_df_gb.sum(axis=1), axis=0)
top_10_df_gb_res = top_10_df_gb.div(top_10_df_gb.sum(axis=1), axis=0)
plot_10 = top_10_df_gb_res.plot(linestyle='-', marker='o', kind = 'line')
plot_10.legend(bbox_to_anchor=(1.01, 1.025))
plot_5 = top_5_df_gb_res.plot(linestyle='-', marker='o', kind = 'line')
plot_5.legend(bbox_to_anchor=(1.01, 1.025))
```

```python
plot_10 = top_10_df_gb.plot(linestyle='-', marker='o', kind = 'line')
plot_10.legend(bbox_to_anchor=(1.01, 1.025))
```

```python
plot_10 = top_10_df_gb_res.plot(linestyle='-', marker='o', kind = 'line')
plot_10.legend(bbox_to_anchor=(1.01, 1.025))
```

```python
plot_5 = top_5_df_gb.plot(linestyle='-', marker='o', kind = 'line')
plot_5.legend(bbox_to_anchor=(1.01, 1.025))
```

```python
plot_5 = top_5_df_gb_res.plot(linestyle='-', marker='o', kind = 'line')
plot_5.legend(bbox_to_anchor=(1.01, 1.025))
```
